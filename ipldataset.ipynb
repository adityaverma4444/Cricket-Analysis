```python
batfirst=ipl[ipl['win_by_runs']!=0]              #all teams theat batted first in first in inning
batfirst.head()
```

```python
plt.figure(figsize=(5,5))
plt.hist(batfirst['win_by_runs'])
plt.xlabel('runs')
plt.title('distributn of win by run')
plt.show()
```

```python
batfirst['winner'].value_counts()
```

```python
plt.figure(figsize=(5,5))     # barplot forteams top3 teams
plt.bar(list(batfirst['winner'].value_counts()[0:3].keys()),batfirst['winner'].value_counts()[0:3])
plt.show()
```

```python
plt.pie(batfirst['winner'].value_counts(),labels=batfirst['winner'].value_counts().keys(),autopct='%0.1f%%')
plt.show()                ## first para - numerical 
```

```python
batsec=ipl[ipl['win_by_wickets']!=0]              #all teams theat batted 2nd
batsec.head()
```

```python
plt.figure(figsize=(5,5))
plt.hist(batsec['win_by_wickets'],bins=25)
plt.xlabel('wicket')   
plt.ylabel('no of matches')
plt.title('distributn of win by wicket')
plt.show()
```

```python
batsec['winner'].value_counts()
```

```python
#create barplot for top 3 team
plt.figure(figsize=(5,5))
plt.bar(list(batsec['winner'].value_counts()[0:3].keys()),list(batsec['winner'].value_counts()[0:3]))
plt.show()
```

```python
#pichart for mot win
plt.pie(batsec['winner'].value_counts()[0:3],labels=batsec['winner'].value_counts()[0:3].keys())
plt.show()
```
